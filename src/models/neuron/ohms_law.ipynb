```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class OhmsLaw():
    """The current in a medium is proportional to the 
    
    I = delta(V)/R
    """
    
    def __init__(self, *, current=None, voltage_difference=None, resistance=None):
        self.i = current
        self.v = voltage_difference
        self.r = resistance
        
    @property
    def current(self):
        """Calculate current
        
        I = V/R
        """
        assert (self.v and self.r), "Missing required data to calculate current"
        
        return self.v / self.r
    
    @property
    def voltage_difference(self):
        """Calculate voltage
        
        V = I*R
        """
        assert (self.i and self.r), "Missing required data to calculate voltage"
        
        return self.i * self.r
    
    @property
    def resistance(self):
        """Calculate resistance
        
        R = V/I
        """
        assert (self.v and self.i), "Missing required data to calculate resistance"
        
        return self.v / self.i
```

```python
ohm = OhmsLaw(voltage_difference=3, resistance=6)

print(f'Current = {ohm.current} amps')
```

```python
terminal1 = 100
terminal2 = 30
voltage_difference = terminal1 - terminal2

resistance_values = np.arange(1, 100, 5).tolist()
```

```python
current_values = [OhmsLaw(voltage_difference=voltage_difference, resistance=r).current for r in resistance_values]
plt.figure(figsize=(15,10))
plt.xlabel('Resistance (Ohms)')
plt.ylabel("Current (Amps)")
plt.plot(resistance_values, current_values)
```
